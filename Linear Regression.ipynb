```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("car.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Engine HP'].fillna(data['Engine HP'].mean(axis=0),inplace=True)
```

```python
data['Engine Fuel Type'].fillna(data['Engine Fuel Type'].mode()[0],axis=0,inplace=True)
```

```python
data.drop('Market Category',axis=1,inplace=True)
```

```python
data['Engine Cylinders'].fillna(data['Engine Cylinders'].mode()[0],axis=0,inplace=True)
```

```python
data['Number of Doors'].fillna(data['Number of Doors'].mode()[0],axis=0,inplace=True)
```

```python
num_data=data.select_dtypes(include=('int64','float64'))
num_data.head()
```

```python
cat_data=data.select_dtypes(exclude=('int64','float64'))
cat_data.head()
```

```python
from sklearn import preprocessing as p
```

```python
le=p.LabelEncoder()
```

```python
for x in cat_data.columns:
    cat_data[x]=le.fit_transform(cat_data[x])
```

```python
cat_data.head()
```

```python
data1=pd.concat([cat_data,num_data],axis=1)
data1.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data1.iloc[:,0:14],data['MSRP'],test_size=0.3,random_state=1)
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score
linreg=linear_model.LinearRegression(normalize=True)
```

```python
linreg.fit(x_train,y_train)
```

```python
prediction=linreg.predict(x_test)
```

```python
print(r2_score(y_test,prediction))
```
